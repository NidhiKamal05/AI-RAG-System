```python
def clean_text(text) :
        text = text.replace("\r", "")
        cleaned_lines = []
        for line in text.split("\n") :
            stripped_line = line.strip()
            if stripped_line :
                cleaned_lines.append(stripped_line)
        text = "\n".join(cleaned_lines)
        return text
```
