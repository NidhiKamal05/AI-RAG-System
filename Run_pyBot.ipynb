```python
import os
import asyncio
from ngrok import ngrok
import nest_asyncio
from dotenv import load_dotenv
load_dotenv()

# Apply nest_asyncio to allow asyncio to run inside an existing event loop
nest_asyncio.apply()

ngrok.set_auth_token(os.getenv("NGROK_AUTH_TOKEN"))

async def get_ngrok_public_url(port):
    # Disconnect any existing tunnels first
    await ngrok.disconnect()
    # Start a new ngrok listener
    listener = await ngrok.connect(port)
    return listener.url()

# Now, call the async function to get the public URL
public_url = asyncio.run(get_ngrok_public_url(5000))
print(f"ngrok tunnel is active at: {public_url}")
```
